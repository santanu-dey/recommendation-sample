```python
import kaggle
# setting up SageMaker parameters

kaggle.api.dataset_download_files('rounakbanik/the-movies-dataset', 
                                  path='./data', 
                                  unzip=True)

print('Completed downloading all files')
```

```python
import os
import json
import sagemaker

sgmk_session = sagemaker.Session()
bucket_name = 'personalize0098900'
bucket_prefix = 'data-staging-area'  # Location in the bucket to store our files


for root, dirs, filenames in os.walk('./data/'):
    for fileName in filenames:
        print ("Uploading "+fileName+" at:")
        file_uri = sgmk_session.upload_data(
        path="./data/"+fileName,
        bucket=bucket_name,
        key_prefix=bucket_prefix)
        print(file_uri)
    break   #prevent descending into subfolders

```
